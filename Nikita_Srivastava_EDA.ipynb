```python
# Converting 'SignupDate' and 'TransactionDate' columns to datetime format
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'])
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])
```

```python
# Merging transactions with customers data based on CustomerID
transactions_customers = pd.merge(transactions, customers, on='CustomerID', how='left')

# Merging the previous result with products data based on ProductID
full_data = pd.merge(transactions_customers, products, on='ProductID', how='left')
```

```python
# Checking the overview of the merged full_data to see the columns and data types
full_data.info()
```

```python
# Getting summary statistics for the full dataset, including both numerical and categorical columns
full_data.describe(include='all')
```

```python
# Plotting the distribution of customers across different regions using a bar plot
region_counts = customers['Region'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=region_counts.index, y=region_counts.values, palette="viridis")
plt.title('Region-wise Customer Distribution')
plt.xlabel('Region')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Analyzing and plotting the trends of customer signups over time (monthly)
signup_trends = customers['SignupDate'].dt.to_period('M').value_counts().sort_index()
signup_trends.plot(kind='line', marker='o', figsize=(8, 5), title='Customer Signup Trends Over Time')
```

```python
# Plotting the top 10 products by total sales value
top_products = full_data.groupby('ProductName')['TotalValue'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_products.values, y=top_products.index, palette="coolwarm")
plt.title('Top 10 Products by Sales Value')
plt.xlabel('Sales Value (USD)')
plt.ylabel('Product Name')
plt.show()
```

```python
# Plotting the distribution of product prices using a box plot
products['Price'].plot(kind='box', vert=False, figsize=(8, 3), title='Product Price Distribution')
```

```python
# Plotting the monthly revenue trend over time
revenue_trend = full_data.groupby(full_data['TransactionDate'].dt.to_period('M'))['TotalValue'].sum()
revenue_trend.index = revenue_trend.index.to_timestamp()

plt.figure(figsize=(10, 6))
revenue_trend.plot(kind='line', marker='o', color='teal')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue (USD)')
plt.grid(True)
plt.show()
```

```python
# Plotting the sales distribution by product category using a pie chart
category_sales = full_data.groupby('Category')['TotalValue'].sum()
plt.figure(figsize=(8, 5))
category_sales.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Sales by Product Category')
plt.ylabel('')
plt.show()
```

```python
# Identifying high-value transactions (top 5% based on revenue)
high_value_transactions = full_data[full_data['TotalValue'] > full_data['TotalValue'].quantile(0.95)]
print(high_value_transactions[['TransactionID', 'CustomerID', 'TotalValue']])
```

```python
# Analyzing seasonal or monthly trends by counting transactions per month
monthly_transactions = full_data['TransactionDate'].dt.to_period('M').value_counts().sort_index()
monthly_transactions.plot(kind='bar', figsize=(10, 5), color='green', title='Transactions by Month')
```

```python
# Identifying repeat customers (those who made more than one purchase)
repeat_customers = full_data['CustomerID'].value_counts()
repeat_customers = repeat_customers[repeat_customers > 1]
print(f"Number of Repeat Customers: {len(repeat_customers)}")

```

```python
# Identifying top 10 customers based on their average purchase value
avg_purchase = full_data.groupby('CustomerID')['TotalValue'].mean().sort_values(ascending=False)
avg_purchase.head(10).plot(kind='bar', figsize=(8, 5), color='purple', title='Top 10 Customers by Avg Purchase Value')
```

```python
# Identifying outliers in customer lifetime value (CLV) by plotting total revenue per customer
customer_revenue = full_data.groupby('CustomerID')['TotalValue'].sum()
sns.boxplot(customer_revenue, color='teal')
plt.title('Customer Lifetime Value Outliers')
```

```python
# Calculating the customer retention rate (percentage of repeat customers)
repeat_customers_rate = len(repeat_customers) / customers['CustomerID'].nunique()
print(f"Customer Retention Rate: {repeat_customers_rate:.2%}")

# Calculating the average order value (AOV) across all transactions
aov = full_data['TotalValue'].mean()
print(f"Average Order Value: ${aov:.2f}")

# Displaying the top 10 customers by lifetime value (total revenue from each customer)
print("Top 10 Customers by Lifetime Value:")
print(customer_revenue.sort_values(ascending=False).head(10))
```

```python
# Rename columns for clarity
full_data.rename(columns={'Price_x': 'TransactionPrice', 'Price_y': 'ProductPrice'}, inplace=True)
```

```python
# Calculating correlations between numerical columns (TransactionPrice, ProductPrice, Quantity, and TotalValue)
correlation_matrix = full_data[['TransactionPrice', 'ProductPrice', 'Quantity', 'TotalValue']].corr()

# Visualizing the correlation matrix with a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap')
plt.show()

```

```python
# Calculating total revenue by multiplying TransactionPrice and Quantity
total_revenue = full_data['TransactionPrice'] * full_data['Quantity']
print(f"Total Revenue: ${total_revenue.sum():,.2f}")

```
